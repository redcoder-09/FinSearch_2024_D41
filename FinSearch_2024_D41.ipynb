```python
import numpy as np
import random
from collections import deque
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

def create_state(data, t, window_size, capital, holding):
    # Extract price and volume data for the state
    price_volume_state = data[t-window_size:t].values.flatten()
    # Append capital and holding to the state
    state = np.append(price_volume_state, [capital, holding])
    return state

# Define parameters
window_size = 10
state_size = window_size * df_scaled.shape[1]  # Number of features in data
action_size = 7  # 0: Hold, 1: Invest 30%, 2: Invest 60%, 3: Invest 90%, 4: Sell 30%, 5: Sell 60%, 6: Sell 90%

class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size + 2  # Adding 2 for capital and holding
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = 0.95    # discount rate
        self.epsilon = 1.0   # exploration rate
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.model = self._build_model()

    def _build_model(self):
        model = Sequential()
        model.add(Dense(128, input_dim=self.state_size, activation='relu'))
        model.add(Dense(128, activation='relu'))
        model.add(Dense(64, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        act_values = self.model.predict(state)
        return np.argmax(act_values[0])  # returns action

    def replay(self, batch_size):
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            # scaled_reward = scale_reward(reward)  # Scale the reward
            target = reward
            if not done:
                target = (reward + self.gamma * np.amax(self.model.predict(next_state)[0]))
            target_f = self.model.predict(state)
            target_f[0][action] = target
            self.model.fit(state, target_f, epochs=1, verbose=0)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def execute_action(self, action, capital, holding, price, old_price):
        initial_portfolio_value = capital + holding * old_price
        price_change = price - old_price
        if action == 0:  # Hold
            return capital, holding, 0, initial_portfolio_value
        elif action == 1:  # Invest 30% of capital
            investment = 0.3 * capital
            holding += investment / price
            capital -= investment
        elif action == 2:  # Invest 60% of capital
            investment = 0.6 * capital
            holding += investment / price
            capital -= investment
        elif action == 3:  # Invest 90% of capital
            investment = 0.9 * capital
            holding += investment / price
            capital -= investment
        elif action == 4:  # Sell 30% of holding
            sell_amount = 0.3 * holding
            capital += sell_amount * price
            holding -= sell_amount
        elif action == 5:  # Sell 60% of holding
            sell_amount = 0.6 * holding
            capital += sell_amount * price
            holding -= sell_amount
        elif action == 6:  # Sell 90% of holding
            sell_amount = 0.9 * holding
            capital += sell_amount * price
            holding -= sell_amount

        # Ensure capital is not negative
        if capital < 0:
            capital = 0
        # Calculate the new portfolio value
        new_portfolio_value = capital + holding * price
        reward = new_portfolio_value - initial_portfolio_value

        return capital, holding, reward, new_portfolio_value

# Initialize the DQN agent
agent = DQNAgent(state_size=state_size, action_size=action_size)

# Training parameters
episodes = 100  # Increased episodes for better learning
batch_size = 64  # Increased batch size for stability
reward_arr = []

# Initialize capital and holding
initial_capital = 100000  # initial capital
initial_holding = 0  # Initial holding


for e in range(episodes):
    print(f"Episode {e+1}/{episodes}")
    capital = initial_capital
    holding = initial_holding
    total_reward = 0
    for time in range(window_size, len(df_scaled) - 1):
        state = create_state(df_scaled, time, window_size=window_size, capital=capital, holding=holding)
        state = np.reshape(state, [1, state_size + 2])
        action = agent.act(state)
        capital, holding, reward, net_worth = agent.execute_action(action, capital, holding, df_scaled.iloc[time  + 1]['Close'], df_scaled.iloc[time]['Close'])

        next_state = create_state(df_scaled, t=time + 1, window_size=window_size, capital=capital, holding=holding)
        next_state = np.reshape(next_state, [1, state_size + 2])
        total_reward += reward
        done = (time == len(df_scaled) - 2)
        agent.remember(state, action, reward, next_state, done)
        state = next_state
        if done:
          print(f"Total Reward: {total_reward}")
          reward_arr.append(total_reward)
          break
    if len(agent.memory) > batch_size:
        agent.replay(batch_size)

# Plot rewards
plt.plot(range(episodes), reward_arr, label='Total Reward per Episode')
plt.xlabel('Episodes')
plt.ylabel('Reward')
plt.legend()
plt.show()

```
